```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('F:\datasets\Wine.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(df)
x
```

```python
pca=PCA(n_components=3)
x=pca.fit_transform(x)
x=pd.DataFrame(x,columns=['PC1','PC2','PC3'])
x
```

```python
from scipy.cluster import hierarchy
```

```python
he=hierarchy.linkage(x,method='complete')
dedo=hierarchy.dendrogram(he)
plt.hlines(y=7,xmin=1,xmax=2111)
```

# from above dendrogram we can tell that their is 4 cluster

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans()
```

```python
wcss=[]
for i in range(2,20):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,20),wcss)
plt.vlines(x=3,colors='red',linestyles=':',ymin=2,ymax=1600)
plt.xlabel('n_cluster')
plt.ylabel('wcss')
plt.title('Elbow graph')
```

From above graph we can conclude that by kmeans method thier is 3 cluster
