```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df_train=pd.read_csv('F:\datasets\SalaryData_Train.csv')
df_train
```

```python
df_test=pd.read_csv('F:\datasets\SalaryData_Test.csv')
df_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
df=pd.concat([df_train,df_test],axis=0)
```

```python
df=df.reset_index()
df
```

```python
df.drop(['index'],inplace=True,axis = 1)
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = df.apply(LabelEncoder().fit_transform)
df
```

```python
df.info()
```

```python
x = df.iloc[:,:13]
y = df['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
nb = MultinomialNB()

nb.fit(xtrain,ytrain)
ypred = nb.predict(xtest)

print(nb.score(xtrain,ytrain))
print(nb.score(xtest,ytest))
print(classification_report(ytest,ypred))
```

```python
from sklearn.naive_bayes import CategoricalNB as CNB
```

```python
CNB = CNB()
CNB.fit(xtrain,ytrain)
ypred = CNB.predict(xtest)

print(CNB.score(xtrain,ytrain))
print(CNB.score(xtest,ytest))
print(classification_report(ytest,ypred))
```

```python
from sklearn.naive_bayes import GaussianNB as GB
```

```python
GB = GB()
GB.fit(xtrain,ytrain)
ypred = GB.predict(xtest)

print(GB.score(xtrain,ytrain))
print(GB.score(xtest,ytest))
print(classification_report(ytest,ypred))
```

# categarical NB has maximum value of accuracy
